## Create a scatter plot

### Import matplotlib library

```python
import matplotlib.pyplot as plt
```

### Access column data separately

```python
df_gapminder.life_expectancy
```

```python
df_gapminder.money_per_person
```

### Pass data to scatter function

```python
plt.scatter(x = df_gapminder. money_per_person, y =  df_gapminder.life_expectancy)
```

## Customize the plot

### Change the size

```python
plt.scatter(x = df_gapminder. money_per_person, y =  df_gapminder.life_expectancy, s=40)
```

### Change the color

```python
plt.scatter(x = df_gapminder. money_per_person, y =  df_gapminder.life_expectancy , c = 'red')
```

### Change the marker

```python
plt.scatter(x = df_gapminder. money_per_person, y =  df_gapminder.life_expectancy, marker= 'x' )
```

### Change the border

```python
plt.scatter(x = df_gapminder. money_per_person, y =  df_gapminder.life_expectancy, marker= 'x', edgecolors='black')
```
